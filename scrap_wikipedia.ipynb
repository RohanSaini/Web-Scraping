```python
#importing beautifulsoup for web sracping and urllib to get the data from web
from urllib.request import urlopen as ureq
from bs4 import BeautifulSoup as soup
```

```python
search_item=input()
```

```python
my_url='https://en.wikipedia.org/wiki/'+search_item
```

```python
client=ureq(my_url)
```

```python
page_html=client.read()
client.close()
#dont forget to close the connection with the website
```

```python
page_soup=soup(page_html,"html.parser")
```

#### If we look at page_soup it is just a html page code now we have to extract the required information out of it

```python
page_soup
```

## there are two ways to find data in this html Find_all and find

```python
find_all_div=page_soup.find_all("div")
find_all_div
```

#### find_all_div gave us all the div in the html code

```python
find_div=page_soup.find("div")
find_div
```

#### find function just give the first occurance of the div

## like this you can find any tag in the html code even with the class name or id name

## This is the output of the search data from wikipedia only first two paragraph

```python
container=page_soup.find("div",{"class":"mw-parser-output"})
container=container.find_all("p")
print(container[0].text)
print(container[1].text)
```
